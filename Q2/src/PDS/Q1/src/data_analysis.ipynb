```python
data_frame = pd.read_csv('../data_raw/Frailty.csv')
print(data_frame.to_string())
```

```python
data_frame.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# the below visualization illustrates the relation between the gripstrength and weight and how it varies on weight.

ax = data_frame.plot.scatter(x='Weight',y='Grip Strength')
fig = ax.get_figure()
fig.savefig("../results/GripStrength_Weight_Relation.png")
```

```python
# the below graph illustrstes the distribution of Gripstrength in the given data

ax = sns.distplot(data_frame['Grip Strength'], kde = False, color = 'red', bins = 30)
fig = ax.get_figure()
fig.savefig("../results/Grip_Strength_Distribution.png")
```
